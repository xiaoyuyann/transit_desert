## Taxi Data

```python
df_taxi = pd.read_csv('Dataset/Taxi_Trips__2013-2023__20240515.csv')
```

```python
df_taxi.columns
```

```python
df_taxi.shape
```

```python
df_taxi['Trip Start Timestamp'] = pd.to_datetime(df_taxi['Trip Start Timestamp'])
df_taxi['Trip End Timestamp'] = pd.to_datetime(df_taxi['Trip End Timestamp'])
df_taxi['Trip Start Timestamp'].describe
```

```python
df_taxi 
```

```python
# output the type of 'Trip Start Timestamp' 
print(type(df_taxi['Trip Start Timestamp'][0]))
# for the 'Trip Start Timestamp' store the rows in df_taxi_2023 where it contains '2023' 
df_taxi_2023 = df_taxi[df_taxi['Trip Start Timestamp'].dt.year == 2023]
# print the first 5 rows of the df_taxi_2023
print(df_taxi_2023.head())
```

```python
# generate a new column to store the year of the trip 'year' from 'Trip Start Timestamp' by extracting the 7th - 11th character
df_taxi['year'] = df_taxi['Trip Start Timestamp'].astype(str).str[6:10]
df_taxi['month'] = df_taxi['Trip Start Timestamp'].astype(str).str[0:2]
df_taxi['day'] = df_taxi['Trip Start Timestamp'].astype(str).str[3:5]
```

```python
df_taxi['year'].head()
df_taxi['month'].head()
df_taxi['day'].head()
```

```python
df_taxi['year'].unique()
```

```python
df_taxi_2022 = df_taxi[df_taxi['year'] == '2017']
df_taxi_2022.head()
```

```python
#df_taxi['Trip Start Timestamp']
# only keep the rows where the first four str = '20230 in 'Trip Start Timestamp'
df_taxi_2023 = df_taxi[df_taxi['Trip Start Timestamp'].astype(str).str[6:10] == '2023']
# df_taxi = df_taxi[df_taxi['Trip Start Timestamp'].dt.year == 2023] 
```

```python
# delete the rows with Trip Miles = 0
df_taxi = df_taxi[df_taxi['Trip Miles'] != 0]
```

```python
# draw histogram of Trip Miles, make the x range from 0 to 100
df_taxi['Trip Miles'].hist(bins=100, range=[0,25])
# x axis 'Trip Miles' y axis 'Frequency'
plt.xlabel('Trip Miles')
plt.ylabel('Frequency')
plt.title('Histogram of Trip Miles of Taxi Trips in Chicago 2015-2023')   
plt.show()
```

```python
df_taxi = df_taxi[df_taxi['Trip Seconds'] != 0]
```

```python
# draw histogram of Trip Seconds, make the x range from 0 to 100
df_taxi['Trip Seconds'].hist(bins=100, range = [0, 6000])
# x axis 'Trip Miles' y axis 'Frequency'
plt.xlabel('Trip Seconds')
plt.ylabel('Frequency')
plt.title('Histogram of Trip Seconds of Taxi Trips in Chicago 2015-2023')   

# plot the vertical line of the mean value of Trip Seconds
plt.axvline(df_taxi['Trip Seconds'].mean(), color='red', linestyle='dashed', linewidth=1)
# show the mean value of Trip Seconds on the middle of the plot
plt.text(df_taxi['Trip Seconds'].mean() +200, 30000, 'Mean = {:.2f}'.format(df_taxi['Trip Seconds'].mean()))
plt.show()
```

```python
# make the box plots for numerical variables in the df_taxi 

import matplotlib.pyplot as plt

# Specify the numerical columns in the DataFrame
numerical_columns = ['Trip Seconds', 'Trip Miles']

# Create a box plot for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.boxplot(df_taxi[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Value')
    plt.show()
```
